```python
# Captures the Most Up-to-Date Health Status
#including multiple responses from the same individual would:
#Violate independence assumptions
#Bias aggregate statistics
#Without filtering, frequent responders count multiple times
#This inflates symptoms, exposure, or vulnerability rates
#Results no longer represent the population accurately
"""
Filtering to the most recent response ensures one observation per participant, prevents over-representation bias, 
and reflects the most current health status—making the data suitable for population-level analysis.
"""
#covid_file1_df = covid_file1_df[covid_file1_df["is_most_recent"] == 'y']
covid_file2_df = covid_file2_df[covid_file2_df["is_most_recent"] == 'y']
```

```python
covid_file2_df.info()
```

```python
#Created date column from week for time based analysis 

covid_file1_df["iso_year"] = 2020
covid_file1_df["date_from_week"] = pd.to_datetime(
    covid_file1_df["iso_year"].astype(str)
    + "-W"
    + covid_file1_df["week"].astype(str).str.zfill(2)
    + "-1",
    format="%G-W%V-%u"
)
covid_file1_df
```

```python
# Check if 'probable' is 'y' AND all specified columns are equal to 'n'
no_symptoms_cols = [
    "fever_chills_shakes", "cough", "shortness_of_breath", 
    "travel_outside_canada", "contact_with_illness"
]
#condition1 = covid_file1_df["probable"] == 'y'
is_no_symptoms = covid_file2_df[no_symptoms_cols].eq('n').all(axis=1)

filtered_df = covid_file2_df[(covid_file2_df["probable"] == 'y') & is_no_symptoms]
filtered_df
```
