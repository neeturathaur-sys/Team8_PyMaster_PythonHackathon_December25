```python
final_df = pd.concat([combined_df, schema_3], ignore_index=True)#combining schema 3
final_df
#final_df['month']
```

```python
# Create a dictionary for mapping weeks to months
week_map = {13: 'March', 14: 'April', 15: 'April', 16: 'April'}

# Apply the mapping to fill NaN values
final_df['month'] = final_df['month'].fillna(final_df['week'].map(week_map))
# Find the most common month (the mode)
most_common_month = final_df['month'].mode()[0]

# Fill NaNs with that value
final_df['month'] = final_df['month'].fillna(most_common_month)
#final_df['month']
```

```python
final_df.columns
```

```python
final_df['age_1'] = final_df['over_60'].map({'y': '>65', 'n': '<65'})
final_df = final_df.drop(columns=['over_60'])
final_df#after dropping over 60 column
```

```python
binary_cols = [
    'probable', 'vulnerable', 'is_most_recent',
       'fever_chills_shakes', 'cough', 'shortness_of_breath',
       'any_medical_conditions', 'travel_outside_canada',
       'contact_with_illness', 'contact_in_household',
       'tested', ]

for col in binary_cols:
    if col in final_df.columns:
      final_df[col] = final_df[col].map({'y': 1, 'n': 0}).fillna(0).astype(int)
final_df# changing y to '0" and N to 1
```

```python
cols = ['month'] + [c for c in final_df.columns if c != 'month']
final_df = final_df[cols]
final_df #moving month to front
```

```python
final_df.isnull().sum()
#print(final_df[['month', 'age_1', 'cough', 'ethnicity']].isnull().sum())
```

```python
# 1. Define groups for different filling strategies
none_fill_cols = [
    'needs', 'mental_health_impact', 'travel_work_school', 
    'self_isolating', 'media_channels', 'financial_obligations_impact', 
    'tobacco_usage'
]

unknown_fill_cols = ['sex', 'age_1']

# 2. Apply the fills
# Fill behavioral/impact columns with 'none'
for col in none_fill_cols:
    if col in final_df.columns:
      final_df[col] = final_df[col].fillna('none')

#3. Fill demographic columns with 'unknown'
for col in unknown_fill_cols:
    if col in final_df.columns:
       final_df[col] = final_df[col].fillna('unknown')

#4.Replace NaN with a label (recommended for reporting)
final_df['covid_results_date'] = final_df['covid_results_date'].fillna('Not Available')

#5. Replace Nan with a label 'Non Tested'
final_df['covid_positive'] = final_df['covid_positive'].fillna('Not Tested')
text_cols = ['symptoms', 'conditions', 'ethnicity', 'needs']

#6.Fill NaNs with 'none' to represent a negative or neutral response
for col in text_cols:
    if col in final_df.columns:
       final_df[col] = final_df[col].fillna('none')
final_df

```

```python
final_df.isnull().sum()
```

```python
final_df.drop('week', axis=1, inplace=True)#dropping week
final_df#after replacing NaN values
```

```python
#A composite symptom variable was created by combining fever/chills, cough, and shortness of breath to indicate overall symptom presence.

# fill NaN values with 0
final_df[['fever_chills_shakes', 'cough', 'shortness_of_breath']] = (
    final_df[['fever_chills_shakes', 'cough', 'shortness_of_breath']].fillna(0)
)

# create combined symptom column
final_df['symptom_cols'] = (
    final_df['fever_chills_shakes'] +
    final_df['cough'] +
    final_df['shortness_of_breath']
)

# convert to binary (1 = has symptoms, 0 = no symptoms)
final_df['symptom_cols'] = (final_df['symptom_cols'] > 0).astype(int)
final_df 
#Treats missing values as no symptom

#Adds the three symptom columns together

#Creates one combined column:

#1 → at least one symptom

#0 → no symptoms
```
