Standardize binary columns.
Our data has many yes/no fields.

Convert them to 0/1 (clean & analytical):

```python
binary_cols = [
    'probable', 'vulnerable', 'is_most_recent',
    'fever_chills_shakes', 'cough',
    'shortness_of_breath', 'over_60',
    'any_medical_conditions', 'travel_outside_canada'
]

df1_clean[binary_cols] = df1_clean[binary_cols].replace({'y': 1, 'n': 0})
```

This enables:

correlations

ML models

clean aggregations

Check missing values

```python
df1_clean.isna().sum().sort_values(ascending=False)
```
